```python
import os
import tarfile
from six.moves import urllib
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# add local path for Georgia Covid-19 data
DATA_PATH = 'C:/Users/Vineet/Documents/georgia_covid_19/datasets/'
```

```python
def load_county_cases(data_path=DATA_PATH):
    csv_path = os.path.join(data_path, "countycases.csv")
    return pd.read_csv(csv_path)
def load_deaths(data_path=DATA_PATH):
    csv_path = os.path.join(data_path, "deaths.csv")
    return pd.read_csv(csv_path)
def load_demographics(data_path=DATA_PATH):
    csv_path = os.path.join(data_path, "demographics.csv")
    return pd.read_csv(csv_path)
def load_county_income(data_path=DATA_PATH):
    csv_path = os.path.join(data_path, "county_income.csv")
    df = pd.read_csv(csv_path)
    df = df.replace(',','', regex=True)
    return df.replace({'$':''}, regex=True)
def load_county_area(data_path=DATA_PATH):
    csv_path = os.path.join(data_path, "county_areas.csv")
    df = pd.read_csv(csv_path)
    df = df.replace(',','', regex=True)
    return df.replace({'$':''}, regex=True)
```

```python
county_cases = load_county_cases()
deaths = load_deaths()
demo = load_demographics()
county_income = load_county_income()
county_area = load_county_area()
```

```python
county['Per capita income'] = [int(x.strip('$').strip(',')) for x in county['Per capita income']]
county['Median household income'] = [int(x.strip('$').strip(',')) for x in county['Median household income']]
county['Median family income'] = [int(x.strip('$').strip(',')) for x in county['Median family income']]
county['Population'] = [int(x) for x in county['Population']]
county['Per capita deaths'] = county['DEATHS']/county['Population']
county['pop_density'] = county['area']/(county['Population']) 
county.head()
```

```python
import numpy as np
def split_train_test(data, test_ratio):
    shuffled_indices = np.random.permutation(len(data))
    test_set_size = int(len(data) * test_ratio)
    test_indices = shuffled_indices[:test_set_size]
    train_indices = shuffled_indices[test_set_size:]
    return data.iloc[train_indices], data.iloc[test_indices]
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import mean_squared_error
encoder = LabelEncoder()
loc_county = county.drop('county_resident', axis=1).drop('Per capita deaths', axis=1).drop('HOSPITALIZATION', axis=1).drop('Positive', axis=1)
loc_county.head()
```

# Cross Validation Splits

```python
county_train_set, county_test_set = split_train_test(loc_county, 0.2)
Y_train = county_train_set['DEATHS']
Y_test = county_test_set['DEATHS']
X_train = county_train_set.drop('DEATHS', axis=1)
X_test = county_test_set.drop('DEATHS', axis=1)
std_Y = Y_test.std()
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X=X_train, y=Y_train)
death_predictions = lin_reg.predict(X_test)
lin_mse = mean_squared_error(Y_test, death_predictions)
lin_rmse = np.sqrt(lin_mse)
lin_rmse/std_Y
```

# Decision Tree Regression

```python
from sklearn.tree import DecisionTreeRegressor
tree_reg = DecisionTreeRegressor()
tree_reg.fit(X=X_train, y=Y_train)
death_predictions = tree_reg.predict(X_test)
lin_mse = mean_squared_error(Y_test, death_predictions)
lin_rmse = np.sqrt(lin_mse)
lin_rmse/std_Y
```

# Random Forest Regression

```python
from sklearn.ensemble import RandomForestRegressor
forest_reg = RandomForestRegressor()
forest_reg.fit(X=X_train, y=Y_train)
death_predictions = forest_reg.predict(X_test)
lin_mse = mean_squared_error(Y_test, death_predictions)
lin_rmse = np.sqrt(lin_mse)
lin_rmse/std_Y
```
